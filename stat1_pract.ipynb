```python
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    
    sns.histplot(sample, kde=True)
    plt.show()
    
    stats.probplot(sample, dist="norm",plot=pylab)
    plt.show()
```

```python
plot_data(log_normal_distribution)
```

```python
plot_data(np.log(log_normal_distribution))
```

```python
#bernouli
p=0.3
sample_size = 1000

bernauli_data = np.random.choice([0,1],size=sample_size,p=[1-p,p])
sns.histplot(bernauli_data, bins=2, kde=False, color="blue")
```

```python
#bionomial distribution
no_of_trials=10
p=0.6  #probability
b_data=np.random.binomial(no_of_trials, p, 10000)
```

```python
sns.distplot(b_data)
```

```python
#piossion distribution

poisson_data = np.random.poisson(lam=5, size=1000)
print(poisson_data)
# plotting a histogram
ax = sns.distplot(poisson_data,
                  kde=False,
                  color='blue')
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')
 
```
